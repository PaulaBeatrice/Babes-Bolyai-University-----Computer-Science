```python
reset()
x,t,w0,x0,v0, = var('x','t','w0','x0', 'v0')
x = function('x')(t)
eqd = diff(x,t,2) + w0^2*x == 0
assume(w0 > 0)
ans1 = desolve (eqd, [x,t])
show(ans1)
```

```python
sol(t,w0,x0,v0)=desolve(eqd, [x,t], [0,x0,v0])
show(sol)
```

```python
plot(sol(t,2,3,4),t,0,10)
```
