```python
reset()
t,s = var('t,s')
f(t,s) = 1 - (s * cos(4*t) * cos(t)) / (sqrt(1 - s**2 * (cos(4 *t))**2 * (sin(t))**2 ))
s = 0.5
plot(f(t,s), 0, 2*pi)
```

```python
reset()
t,s = var('t,s')
f(t,s) = 1 - (s * cos(4*t) * cos(t)) / (sqrt(1 - s**2 * (cos(4 *t))**2 * (sin(t))**2 ))
x(t) = f(t-pi/2, s)
y(t) = f(t,s)

g = parametric_plot( (f(t-pi/2,1/10) , f(t,1/10) ) , (t,0,2*pi) )

for k in [2..10]:
    g1 = parametric_plot( ( f(t-pi/2, k/10) , f(t,k/10) ) , (t,0,2*pi) )
    g = g + g1
show(g)
```
