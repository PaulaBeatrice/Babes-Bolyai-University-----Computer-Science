```python
reset()
x = var('x')
y = function('y')(x)
eqd = diff(y,x) == 2 * x *(1 + y^2)
sol = desolve(eqd,y)
solve(sol,y)
sol_e = solve(sol,y)
sol_e2(x, _C) = sol_e[0].rhs()
plot(sol_e2(x,1),x, 0,10, detect_poles = 'True', ymin = -10,ymax = 10)
```

```python
reset()
x = var('x')
y = function('y')(x)
eqd = (x^2 - 1) * diff(y,x) + 2 * x * y^2 == 0
sol = desolve(eqd,y)
show(solve(sol,y))
sol_e = solve(sol,y)
sol_e2(x, _C) = sol_e[0].rhs()
plot(sol_e2(x,0),x, -2,2, detect_poles = 'True', ymin = -10,ymax = 10)
```

```python
reset()
x = var('x')
y = function('y')(x)
eqd = 2 * x^2 * diff(y,x) == x^2 + y^2
sol = desolve(eqd,y)
show(solve(sol,y))
sol_e = solve(sol,y)
sol_e2(x, _C) = sol_e[0].rhs()
plot(sol_e2(x,1),x, 10,100, detect_poles = 'True', ymin = -10,ymax = 10)
```
